```python
import numpy as np
import pandas as pd

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
datos = pd.read_csv('C:/Users/Nadia/Documents/Especializaci√≥n/NLP/Clases y Talleres/dinosaurios.csv')
datos.head()
```

```python
len(datos)
```

```python
palabras = datos.Dinosaurios.str.cat(sep=" ")
wordcloud = WordCloud().generate(palabras)

plt.imshow(wordcloud)
plt.show()
```

```python
palabras = datos.Dinosaurios.str.cat(sep=' ')

wordcloud = WordCloud(width=800, height=400, # Ajustamos el tama√±o
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='gist_heat',
                     stopwords=stopwords).generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") #apaga el eje para que no se visualice
plt.tight_layout(pad=0)
plt.show()
```

```python
palabras = datos.Dinosaurios.str.cat(sep=' ')

 

wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='copper_r',
                     stopwords=stopwords,
                     mask=mask, contour_width=0.5, contour_color='green').generate(palabras)
wordcloud.to_file("dinosaurio.png")

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
