Punto 1

Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con with open

```python
with open("C:/Users/Nadia/Documents/Especialización/NPL/Clases y Talleres/UNA_SOLITARIA_VOZ_HUMANA.txt", "r",encoding = 'utf-8') as archivo:
    texto = archivo.read()
```

```python
print (texto)
```

Convertir a minúsculas

```python
texto = texto.lower()
```

```python
print (texto)
```

Conservar sólo palabras

```python
import re
```

```python
texto = re.sub(r"[\W\d_]+", " ", texto)
```

```python
print (texto)
```

Tokenizar

```python
token = texto.split()
```

```python
token
```

Punto 2

Quitar Palabras Vacias

```python
import nltk
```

```python
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
palabras = [palabra for palabra in token if palabra not in stopwords_sp]
palabras
```

¿Cuáles son las 10 palabras no vacias más usadas?

```python

import pandas as pd
import collections
conteo_palabras = collections.Counter(palabras)
print(conteo_palabras)

```

 Stemming del documento

```python
from nltk.stem.snowball import SnowballStemmer
englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")
```

```python
spanishStemmer.stem('solitario')
```

```python
stem = []

for palabra in palabras:
    raiz = spanishStemmer.stem(palabra)
    stem.append(raiz)
    
stem
```

 ¿Cuáles son las 10 raíces más usadas?

```python
import collections
conteo_stem = collections.Counter(stem)
print(conteo_stem)
```
